# Data Preprocessing

## Handling  Missing Values

Impute the missing values "mean" for numerical columns
and for categorical columns with its "mode"

```python
# replace 0 with NAN
import numpy as np
df1_copy = df.copy()
df1_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
df1_copy.head()
```

```python
df1_copy['Outcome'].value_counts()
```

```python
print(df1_copy.isnull().sum())
```

```python
#pip install missingno
```

```python
import missingno as msno
msno.bar(df1_copy);
```

```python
df1_copy['Insulin'].fillna(df_copy['Insulin'].mean(), inplace=True)
df1_copy['Glucose'].fillna(df_copy['Glucose'].mean(), inplace=True)
df1_copy['BloodPressure'].fillna(df_copy['BloodPressure'].mean(), inplace=True)
df1_copy['SkinThickness'].fillna(df_copy['SkinThickness'].mean(), inplace=True)
df1_copy['BMI'].fillna(df_copy['BMI'].mean(), inplace=True)
```
